```python
# To find ORFs in all the three frames of '+' strand only in a multifasta file. This code would provide the start codons and stop codons and length of each ORF above 30 nucleotides.

def orf_count(file_name):
    from Bio import SeqIO
    file = open (file_name)
    file.seek(0)
    for record in SeqIO.parse(file_name, "fasta"):
        print (record.id)
        print (record.seq)
        
        # For frame 1
        mystartcodon1 = []
        mystopcodon1 = []
        for i in range (0,len(record.seq)-2,3):
            if record.seq[i:i+3] == "ATG":
                mystartcodon1.append(i)

        for j in range (0,len(record.seq)-2,3):
            if record.seq[j:j+3] == "TAG" or record.seq[j:j+3] == "TAA" or record.seq[j:j+3] == "TGA":
                mystopcodon1.append(j)
            
        finalstartcodons1= []
        finalstopcodons1 = [0]
        for i in mystartcodon1:
            for j in mystopcodon1:
                for k in finalstopcodons1:
                    if i<j and i not in finalstartcodons1 and i> finalstopcodons1[-1]:
                        finalstartcodons1.append(i)
                        finalstopcodons1.append(j)
        
        print ("Final start codons- frame1")
        F1 = 1
        startcodons_frame1 = [ x+F1 for x in finalstartcodons1]
        print (startcodons_frame1)
        
        finalstopcodons1.remove(0)
        print ("Final stop codons- frame1")
        FS1 = 3
        stopcodons_frame1 = [x+FS1 for x in finalstopcodons1]
        print (stopcodons_frame1)
        
        orf_length1 = []
        zip_object1 = zip (finalstopcodons1, finalstartcodons1)
        for finalstopcodons1_i, finalstartcodons1_i in zip_object1:
            orf_length1.append((finalstopcodons1_i+3)-finalstartcodons1_i)
        final_orfs1 = []
        for i in orf_length1:
            if i > 30:
                final_orfs1.append(i)
        print ("orfs length frame 1")
        print(final_orfs1)
        
        # For frame 2
        mystartcodon2 = []
        mystopcodon2 = []
        for a in range (1,len(record.seq)-2,3):
            if record.seq[a:a+3] == "ATG":
                mystartcodon2.append(a)

        for b in range (1,len(record.seq)-2,3):
            if record.seq[b:b+3] == "TAG" or record.seq[b:b+3] == "TAA" or record.seq[b:b+3] == "TGA":
                mystopcodon2.append(b)
                
        finalstartcodons2 = []
        finalstopcodons2 = [0]
        for a in mystartcodon2:
            for b in mystopcodon2:
                for c in finalstopcodons2:
                    if a<b and a not in finalstartcodons2 and a> finalstopcodons2[-1]:
                        finalstartcodons2.append(a)
                        finalstopcodons2.append(b)
        
        print ("Final start codons-frame2")
        F1 = 1
        startcodons_frame2 = [ x+F1 for x in finalstartcodons2]
        print (startcodons_frame2)
       
        
        finalstopcodons2.remove(0)
        print ("Final stop codons-frame2")
        FS1 = 3
        stopcodons_frame2 = [x+FS1 for x in finalstopcodons2]
        print (stopcodons_frame2)
        
        orf_length2 = []
        zip_object2 = zip (finalstopcodons2, finalstartcodons2)
        for finalstopcodons2_a, finalstartcodons2_a in zip_object2:
            orf_length2.append((finalstopcodons2_a+3)-finalstartcodons2_a)
        final_orfs2 = []
        for a in orf_length2:
            if a > 30:
                final_orfs2.append(a)
        print ("Orfs length frame 2")
        print(final_orfs2)
        
        # For frame 3
        mystartcodon3 = []
        mystopcodon3 = []
        for d in range (2,len(record.seq)-2,3):
            if record.seq[d:d+3] == "ATG":
                mystartcodon3.append(d)

        for e in range (2,len(record.seq)-2,3):
            if record.seq[e:e+3] == "TAG" or record.seq[e:e+3] == "TAA" or record.seq[e:e+3] == "TGA":
                mystopcodon3.append(e)
                
        finalstartcodons3 = []
        finalstopcodons3 = [0]
        for d in mystartcodon3:
            for e in mystopcodon3:
                for f in finalstopcodons3:
                    if d<e and d not in finalstartcodons3 and d> finalstopcodons3[-1]:
                        finalstartcodons3.append(d)
                        finalstopcodons3.append(e)
                        
        print ("Final start codons- frame 3")
        F1 = 1
        startcodons_frame3 = [ x+F1 for x in finalstartcodons3]
        print (startcodons_frame3)

        finalstopcodons3.remove(0)
        print ("Final stop codons- frame3")
        FS1 = 3
        stopcodons_frame3 = [x+FS1 for x in finalstopcodons3]
        print (stopcodons_frame3)

        orf_length3 = []
        zip_object3 = zip (finalstopcodons3, finalstartcodons3)
        for finalstopcodons3_d, finalstartcodons3_d in zip_object3:
            orf_length3.append((finalstopcodons3_d+3)-finalstartcodons3_d)
        final_orfs3 = []
        for d in orf_length3:
            if d > 30:
                final_orfs3.append(d)
        print ("Orfs length frame 3")
        print(final_orfs3)
```
